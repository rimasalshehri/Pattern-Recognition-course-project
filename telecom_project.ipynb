```python
one_hot_encoded_data = pd.get_dummies(data, columns = ['state', 'area code'])
```

```python
X=one_hot_encoded_data.copy()
```

```python
X
```

```python
# create training and testing set
from sklearn.model_selection import train_test_split
x_trainset, x_testset, y_trainset, y_testset = train_test_split(X, y, test_size=0.25, random_state=7)
```

```python
x_trainset.shape
```

```python
x_testset.shape
```

```python
y_trainset.shape
```

```python
y_testset.shape
```

```python
#Create Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(x_trainset, y_trainset)
# prediction
predictions = clf.predict(x_testset)
print(predictions[:5])
```

```python
# accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_testset, predictions))
```

```python
#precsion
# Looking at the precision score
from sklearn.metrics import precision_score
print(precision_score(y_testset, predictions))
```

```python
# Looking at the recall score
from sklearn.metrics import recall_score
print(recall_score(y_testset,predictions))
```

```python
# confusion matrics
from sklearn.metrics import confusion_matrix
confusion_matrix(y_testset, predictions)
```

```python
sns.heatmap(confusion_matrix(y_testset, predictions), annot=True)
```

```python
# Creating a dictionary of parameters to use in GridSearchCV
from sklearn.model_selection import GridSearchCV

params = {
    'criterion':  ['gini', 'entropy'],
    'max_depth':  [None, 2, 4, 6, 8, 10],
    'max_features': [None, 'sqrt', 'log2', 0.2, 0.4, 0.6, 0.8],
    'splitter': ['best', 'random']
}

clf = GridSearchCV(
    estimator=DecisionTreeClassifier(),
    param_grid=params,
    cv=5,
    n_jobs=5,
    verbose=1,
)
clf.fit(x_trainset, y_trainset)
print(clf.best_params_)
```

```python
clf = DecisionTreeClassifier(max_depth=6, criterion='gini', max_features=None, splitter='best')
clf.fit(x_trainset,y_trainset)
predictions = clf.predict(x_testset)
print(accuracy_score(y_testset, predictions))
```
